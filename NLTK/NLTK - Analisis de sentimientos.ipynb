```python
from nltk.corpus import twitter_samples
from nltk.tag import pos_tag_sents
```

```python
twitter_samples.fileids()
```

```python
twitter_samples.strings('tweets.20150430-223406.json')
```

```python
#Todos los comentarios positivos
a = twitter_samples.strings('positive_tweets.json')
a_tokens = twitter_samples.tokenized('positive_tweets.json')
a_tokens
```

```python
a_tagged = pos_tag_sents(a_tokens)
a_tagged
#JJ, adjetivos
#NN, nombres en singular
#NNS, nombres en plural
```

```python
JJ_count = 0
NN_count = 0
for tweet in a_tagged:
    for par in tweet:
        tag = par[1]
        if tag == 'JJ':
            JJ_count += 1
        elif tag == 'NN':
            NN_count += 1
print('Qty adjetivos: ', JJ_count)
print('Qty de nombres: ', NN_count)
```
