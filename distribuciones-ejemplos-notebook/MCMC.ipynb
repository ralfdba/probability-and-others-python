```python
#Los metodos de montecarlo se basan en la analogia entre probabilidad y volumen
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
np.random.seed(1984)
```

```python
def mc_pi(N=10000):
    plt.figure(figsize=(9,9))
    x,y = np.random.uniform(-1,1,size=(2,N))
    interior = (x**2 + y**2) <= 1
    exterior = np.invert(interior)
    pi = interior.sum() * 4 / N
    error = abs((pi - np.pi) / pi) * 100
    plt.plot(x[interior],y[interior],'b.')
    plt.plot(x[exterior],y[exterior],'r.')
    plt.plot(0,0,label="$\hat \pi$ = {:4.4f}\nerror = {:4.4f}%".format(pi,error),alpha=0)
    plt.axis("square")
    plt.legend(frameon=True,framealpha=0.8,fontsize=17)

mc_pi()
```

```python
mc_pi(N=100000)
```

```python
mc_pi(N=1000000)
```
