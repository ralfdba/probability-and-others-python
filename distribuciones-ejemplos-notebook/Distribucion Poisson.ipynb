```python
#Poisson, describe la probabilidad de encontrar con exactitud eventos enteros
#en un lapso de tiempo.
#Distribucion discreta, aquellas en que la variable toma valores determinados
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
mu = 3.6 #parametro
poisson = stats.poisson(mu) #Distribucion
x = np.arange(poisson.ppf(0.01),poisson.ppf(0.99))
fmp = poisson.pmf(x)
```

```python
#Graficos
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp,colors='g',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Poisson")
plt.show()
```

```python
#histogramas
numeros = poisson.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Poisson")
plt.show()
```
