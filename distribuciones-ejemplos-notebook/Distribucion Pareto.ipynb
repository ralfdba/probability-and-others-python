```python
#Distribucion de pareto, o conocidad como la regla 80/20, la proabilidad de una distribucion continua entre 2 parametros
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
pareto = stats.pareto(2.3)
x = np.linspace(pareto.ppf(0.01),pareto.ppf(0.99),100)
fp = pareto.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Pareto ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = pareto.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Pareto")
plt.show()
```
