```python
#Dist Exponencial, es una distribucion continua para modelar los tiempos de espera para la ocurrencia de cierto evento conocido
#es similar a la distribucion geometrica
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
exponencial = stats.expon()
x = np.linspace(exponencial.ppf(0.01),exponencial.ppf(0.99),100)
fp = exponencial.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist exponencial ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = exponencial.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma exponencial")
plt.show()
```
