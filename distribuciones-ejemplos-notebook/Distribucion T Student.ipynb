```python
#Distribucion T Student, determina la probabilidad de diferencia entre las medias de dos muestras de observacion
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
t = stats.t(70)
x = np.linspace(t.ppf(0.01),t.ppf(0.99),100)
fp = t.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist T student ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = t.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma T student")
plt.show()
```
