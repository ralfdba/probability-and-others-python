```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
sns.set_palette("deep",desat=.6)
sns.set_context(rc={"figure.figsize":(8,4)})
#configurar el nivel de replicacion
np.random.seed(2016)
```

```python
#histograma de prueba hist()
#md = media, dst = desviacion estandar
md,dst = 0, 0.2
datos = np.random.normal(md,dst,1000)
a,b,c = plt.hist(datos,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Prueba")
plt.show()
```

```python
#FMP, Funcion de Masa de Probabilidad, para las distribuciones discretas
#Otra forma de representar las distribuciones es utilizando FMP
#Relacion de cada valor con su probabilidad y no con la frecuencia.
#utilizar la libreria scipy, la funcion pmf(fmp)
```

```python
n,p = 30, 0.4#parametros vamos a utilizar
n_1,p_1 = 20, 0.3
#arange(start,stop,step)
x = np.arange(stats.binom.ppf(0.01,n,p),stats.binom.ppf(0.99,n,p))
x_1 = np.arange(stats.binom.ppf(0.01,n_1,p_1),stats.binom.ppf(0.99,n_1,p_1))
#creando la funcion de masa de probabilidad (FMP (PMF en ingles))
fmp = stats.binom.pmf(x,n,p)
fmp_1 = stats.binom.pmf(x_1,n_1,p_1)
#Graficar
plt.plot(x,fmp,'--')
plt.plot(x_1,fmp_1)
plt.vlines(x,0,fmp,colors='r',lw=5,alpha=0.5)
plt.vlines(x_1,0,fmp_1,colors='g',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("FMP")
plt.show()
```

```python
#FDA(CDF), se utiliza cuando los valores del set de datos son muy grandes
#Nos sirve para distribuciones discretas y continuas
#Relaciona los valores con su correspondiente percentil
#Va a describir la probabilidad de que una variable X este sometida a una distribucion de
#probabilidad que se situe en la zona de valores menores o iguales a X
```

```python
d = np.linspace(stats.norm(10,1.2).ppf(0.01),stats.norm(10,1.2).ppf(0.99),100)
fda_binomial = stats.binom.cdf(x,n,p)
fda_normal = stats.norm(10,1.2).cdf(d)
#Graficar
plt.plot(x,fda_binomial,'--',label='FDA Binom')
plt.plot(d,fda_normal,label='FDA Normal')
#plt.vlines(x,0,fmp,colors='r',lw=5,alpha=0.5)
#plt.vlines(x_1,0,fmp_1,colors='g',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("FDA")
plt.grid(True)
plt.show()
```

```python
#FDP(pdf)
fdp_normal = stats.norm(10,1.2).pdf(d)
plt.plot(d,fdp_normal,label='FDA Normal')
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("FDP")
plt.grid(True)
plt.show()
```
