```python
#Distribucion Log Normal, se aplica a variables aleatorias limitadas por ceros (0), como la distribucion de asimetria positiva
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
a = 0.6
lnormal = stats.lognorm(a)
x = np.linspace(lnormal.ppf(0.01),lnormal.ppf(0.99),100)
#Funcion de Probabilidad
fp = lnormal.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Log Normal ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = lnormal.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Log Normal")
plt.show()
```
