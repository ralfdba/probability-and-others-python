```python
#Distribucion Hipergeometrica, Q pruebas al azar y la probabilidad de que los elementos
#Provengan del mismo grupo (Q) al azar
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
Q,n,N = 30,10,12
hgeom = stats.hypergeom(Q,n,N)
x = np.arange(0,n+1)
fmp = hgeom.pmf(x)
```

```python
#Graficos
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp,colors='r',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Hipergeometrica")
plt.show()
```

```python
#histogramas
numeros = hgeom.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Hipergeometrica")
plt.show()
```
