```python
#Dist Gamma, es una distribucion entre 2 parametros (es una dist de un caso particular la dist exponencial)
#Tambien es una distribucion continua
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
gamma = stats.gamma(2.6)
x = np.linspace(gamma.ppf(0.01),gamma.ppf(0.99),100)
fp = gamma.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist gamma ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = gamma.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma gamma")
plt.show()
```
