```python
#Distribucion de Bernoulli, que probabilidad (p) hay de que haya exito en un 1 sola prueba
#la probabilidad de fracaso 1 - p
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
p = 0.5
bernoulli = stats.bernoulli(p)
x = np.arange(-1,3)
fmp = bernoulli.pmf(x)
```

```python
#Graficos
fig,ax = plt.subplots()
ax.plot(x,fmp,'bo')
ax.vlines(x,0,fmp,colors='r',lw=5,alpha=0.5)
ax.set_yticks([0.,0.2,0.4,0.6])
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Bernoulli")
plt.show()
```

```python
#histogramas
numeros = bernoulli.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Bernoulli")
plt.show()
```
