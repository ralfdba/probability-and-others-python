```python
#Distribucion Uniforme, es la que todos los valores de las matrices tienen la misma probabilidad de ocurrencia
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
uniforme = stats.uniform()
x = np.linspace(uniforme.ppf(0.01),uniforme.ppf(0.99),100)
#Funcion de Probabilidad
fp = uniforme.pdf(x)
```

```python
#Graficos
fig,ax = plt.subplots()
ax.plot(x,fp,'--')
ax.vlines(x,0,fp,colors='b',lw=5,alpha=0.5)
ax.set_yticks([0.,0.2,0.4,0.8,1.,1.2])
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Uniforme ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = uniforme.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Uniforme")
plt.show()
```
