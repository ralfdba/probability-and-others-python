```python
#Dist Beta, es una distribucion entre 2 parametros (es una dist de un caso particular la dist exponencial)
#Tambien es una distribucion continua, pero definidas por el intervalo (0,1)
#Dist beta se utiliza como el a priori en el analisis bayesiano en proporciones binomiales
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
beta = stats.beta(2.6,0.6)
x = np.linspace(beta.ppf(0.01),beta.ppf(0.99),100)
fp = beta.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist beta ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = beta.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma beta")
plt.show()
```
