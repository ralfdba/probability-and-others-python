```python
#Distribucion Chi Cuadrado o la distribucion de Pearson, es la suma del cuadrado de las variables aleatorias
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
chji2 = stats.chi2(36)
x = np.linspace(chji2.ppf(0.01),chji2.ppf(0.99),100)
fp = chji2.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist chji2 ")
plt.grid(True)
plt.show()
```

```python
#histogramas
numeros = chji2.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma chji2")
plt.show()
```
