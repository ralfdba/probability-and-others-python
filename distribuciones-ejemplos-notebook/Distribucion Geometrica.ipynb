```python
#Distribucion Geometrica, es la probabilidad exito es N cantidad prueba previamente definidas
#Ejemplo RRHH
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#Cantidad de pruebas
N = 0.3
geometrica = stats.geom(N)
x = np.arange(geometrica.ppf(0.01),geometrica.ppf(0.99))
fmp = geometrica.pmf(x)
```

```python
#Graficos
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp,colors='g',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Binomial")
plt.show()
```

```python
#histogramas
numeros = geometrica.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Binomial")
plt.show()
```
