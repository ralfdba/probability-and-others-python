```python
#Dist Binomial, probabilidad de que un entero tenga exito en N cantidad de pruebas
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#N = cantidad de pruebas
#p = probabilidad
N,p = 30, 0.4
binomial = stats.binom(N,p)#Creamos la distribucion
x = np.arange(binomial.ppf(0.01),binomial.ppf(0.99))
fmp = binomial.pmf(x)
```

```python
#Graficos
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp,colors='g',lw=5,alpha=0.5)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Binomial")
plt.show()
```

```python
#histogramas
numeros = binomial.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Binomial")
plt.show()
```
