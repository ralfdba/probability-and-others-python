```python
#Distribucion Normal, es la media aritmetica
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#m = media
#sdt = desviacion estandar
m,sdt = 0,0.2
normal = stats.norm(m,sdt)
x = np.linspace(normal.ppf(0.01),normal.ppf(0.99),100)
fp = normal.pdf(x)
```

```python
#Graficos
plt.plot(x,fp)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.title("Dist Normal|Gauss ")
plt.show()
```

```python
#histogramas
numeros = normal.rvs(1000)
a,b,c = plt.hist(numeros,20)
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma Normal|Gauss")
plt.show()
```
